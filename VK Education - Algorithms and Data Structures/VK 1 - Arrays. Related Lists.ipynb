````python
class Array:
    def __init__(self, capacity: int, el_type: any):
        self.array = [None] * capacity
        self.el_type = el_type
        
    def get(self, index: int):
        return self.array[index]
    
    def append(self, index: int, value: any):
        # Check data type
        if not isinstance(value, self.el_type):
            raise ValueError(f'value is {type(value)}, but must be {self.el_type}')
            
        self.array[index] = value            
````

def append_to_the_end(number_of_marks, marks) -> str:
    positive_marks = [x for x in marks if x > 0]
    
    for _ in range(len(marks) - len(positive_marks)):
        positive_marks.append(0)
    
    positive_marks = map(str, positive_marks)
    
    return " ".join(positive_marks)

````python
from collections import deque

def find_min(dq, inputs):
    for _ in range(inputs):
        # print(dq)
        action = input()
        
        if action[0] == '+':
            if action[1] == 'L':
                dq.appendleft(int(action[3:]))
                print(min(dq))
                
            elif action[1] == 'R':
                dq.append(int(action[3:]))
                print(min(dq))
                
        else:  # action[0] == '-':
            if action[1] == 'L':
                dq.popleft()
                if len(dq):
                    print(min(dq))
                    
                else:
                    print(-1)
                    
            elif action[1] == 'R':
                dq.pop()
                if len(dq):
                    print(min(dq))
                    
                else:
                    print(-1)
            
    
deq = deque(maxlen=1_000_000)
operators = int(input())
find_min(deq, operators)
````

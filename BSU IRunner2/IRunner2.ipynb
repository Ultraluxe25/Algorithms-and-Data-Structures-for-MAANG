````python
from collections import deque

def find_min(dq, inputs):
    for _ in range(inputs):
        # print(dq)
        action = input()
        
        if action[0] == '+':
            if action[1] == 'L':
                dq.appendleft(int(action[3:]))
                print(min(dq))
                
            elif action[1] == 'R':
                dq.append(int(action[3:]))
                print(min(dq))
                
        else:  # action[0] == '-':
            if action[1] == 'L':
                dq.popleft()
                if len(dq):
                    print(min(dq))
                    
                else:
                    print(-1)
                    
            elif action[1] == 'R':
                dq.pop()
                if len(dq):
                    print(min(dq))
                    
                else:
                    print(-1)
            
    
deq = deque(maxlen=1_000_000)
operators = int(input())
find_min(deq, operators)
````

````python
def count_different_bits(x, y):
    bit_size = 64  # Size of the unsigned long long data type in bits
    for bit_position in range(bit_size):
        if x == y:
            return bit_position
        x >>= 1
        y >>= 1

def count_set_bits(x):
    set_bit_count = 0
    bit_size = 64  # Size of the unsigned long long data type in bits
    for bit_position in range(bit_size):
        if x & 1:
            set_bit_count += 1
        x >>= 1
    return set_bit_count

def calculate_bit_difference(x, add):
    y = x + add
    return add * 2 + count_set_bits(x) - count_set_bits(y)

def greatest_common_divisor(a, b):
    while b:
        a %= b
        temp = b
        b = a
        a = temp
    return a

def reduce_fraction(frac):
    gcd = greatest_common_divisor(frac['num'], frac['denum'])
    return {'numerator': frac['num'] // gcd, 'denominator': frac['denum'] // gcd}

def main():
    bit_count = int(input())
    bits = list(map(int, input().split()))  # Read bits as a list of numbers separated by spaces
    x = 0
    for bit_value in bits:
        x = (x << 1) | bit_value

    add_value = int(input())
    
    print(count_different_bits(x, x + add_value))
    frac = {'num': calculate_bit_difference(x, add_value), 'denum': add_value}
    frac = reduce_fraction(frac)
    print(f"{frac['numerator']}/{frac['denominator']}")

if __name__ == "__main__":
    main()
````

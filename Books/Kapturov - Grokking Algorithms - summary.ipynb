# <p style="color: red">Chapter 1: Binary search</p>

```python
def binary_search(array, value):
    start = 0
    end = len(array) - 1
    iterations = 0

    while start <= end:
        middle = (start + end) // 2
        iterations += 1

        if array[middle] == value:
            print(f'Iteration number: {iterations}')
            return f'Index number: {middle}'
        elif array[middle] < value:
            start = middle + 1
        elif array[middle] > value:
            end = middle - 1
            
    return 'There is no searchin number in the array!'


sorted_array = [2, 3, 56, 114, 567, 876, 912, 1000]
target = 912
print(binary_search(sorted_array, target))
```

# <p style="color: red">Chapter 2: Selection Sort</p>

## My 1st varient of the selection sort

```python
def selection_sort(array):  # Ascending order
    n = len(array)
    for i in range(n - 1):
        smallest = i  # First element in each iterations could be the smallest
        for j in range(i + 1, n):
            if array[smallest] > array[j]:
                smallest = j  # Index of the current smallest value

        array[smallest], array[i] = array[i], array[smallest]  # Swap current element of the external with the smallest

    return array


numbers = [5, 18, 1, -31, 0, 23, 81, 39]
print(selection_sort(numbers))
```

## My 2nd varient of the selection sort

```python
def find_smallest(arr):  # Find index of the smallest element
    smallest_element = arr[0]
    smallest_index = 0
    
    for index, element in enumerate(arr):
        if element < smallest_element:
            smallest_index = index
            smallest_element = element
            
    return smallest_index


def select_sort_array(array):
    sorted_array = []
    while array:  # While there are at least one element in array
        sorted_array.append(array.pop(find_smallest(array)))

    return sorted_array


unsorted_array = [18, 27, -243, -394, 15, 29, 1023, 16]
print(select_sort_array(unsorted_array))
```
